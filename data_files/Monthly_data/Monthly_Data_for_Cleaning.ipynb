```python
import pandas as pd
df = pd.read_csv('Monthly_data (Editing).csv')
print(df)
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.dropna()
print(df)
```

```python
missing_values = df.isnull()
print("DataFrame with missing values (NaNs):\n", missing_values)
```

```python
file_path = 'Monthly_data (Cleaned).csv'

df.to_csv(file_path, index=False)  
print(f'Data saved to {file_path}')
```
